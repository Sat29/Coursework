```python

import numpy as np
import matplotlib.pyplot as plt
loaded_array = np.load("londonIncidents.npz")

#get data from file into x and y values
my_dict = {}

for item in loaded_array['all']:
    key = item[0]
    value = item[1]
    my_dict[key] = value

x_values = list(my_dict.keys())
y_values = list(my_dict.values())


# Define the number of grid divisions in both x and y directions
x_bins = 10
y_bins = 10

#bounds for x and y axis
ch_xmin = 0
ch_xmax = 46
ch_ymin = -17.5
ch_ymax = 0

#getting bin size for x and y bins
b_xsize = (ch_xmax - ch_xmin)/x_bins
b_ysize = (ch_ymax - ch_ymin)/y_bins

#Make histogram
hist, x_lines, y_lines = np.histogram2d(x_values, y_values, bins=(x_bins, y_bins))


#Count data points in each grid
grid_counts = {}
for i in range(x_bins):
    for j in range(y_bins):
        count = hist[i, j]
        if count in grid_counts:
            grid_counts[count].append((i+1, j+1))
        else:
            grid_counts[count] = [(i+1, j+1)]

#Show how many grids contain l number of points
for count, grids in grid_counts.items():
    print(f"Number of grids with {count} points: {len(grids)}")

'''
#Counts data points in each cell
for i in range(x_bins):
    for j in range(y_bins):
        count = hist[i, j]
        print(f"Grid ({i+1}, {j+1}): {count} points")
'''

#Showing horizontal and vertical grid lines
for i in range(x_bins):
  x_lines = ch_xmin + i * b_xsize
  plt.axvline(x_lines, color='black', linestyle='-')

for i in range(y_bins):
  y_lines = ch_ymin + i * b_ysize
  plt.axhline(y_lines, color='black', linestyle='-')


#plot
plt.scatter(x_values, y_values)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.title('Scatter Plot with Grid')
plt.show()
```

```python
#Part d code for problem 4.11

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


points = [2,7,5,11,12,8,4,14,25,22,18,9,10,17,20,19,6,13,15,16,21] #number of points in a grid (l in the problem)
num_grid = [1,5,3,8,9,4,3,11,2,2,4,5,10,3,3,3,7,6,5,5,1] #corresponding number of grides with points l above

#Get num_grid as probability which will be our y-value
const = 0.01
prob = []
for i in  range(len(num_grid)):
  prob.append(num_grid[i] * const)


#Get average points value
avgP = sum(points)/len(points)

#get standard deviation of points
std_dev_points = np.std(points)

#Now we can get a guassian distribution for points data
gauss_points = norm.pdf(points,avgP, std_dev_points)


#plot
plt.bar(points, prob, label = 'Guassian Distribution')
plt.scatter(points, gauss_points, color = 'k', label = 'Estimated Distribution')
plt.xlabel('l (Dots in a cell)')
plt.ylabel('P(l)')
plt.legend()
plt.show()
```
